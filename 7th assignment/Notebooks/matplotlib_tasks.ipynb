```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = {
    "Name": ["Aidar", "Arsen", "Daur", "Alik", "Dana"],
    "Score": [75, 88, 92, 81, 67]
}

df = pd.DataFrame(data)
```

```python
plt.plot(df["Name"], df["Score"])
plt.xlabel("Student Name")
plt.ylabel("Score")
plt.title("Student Scores Line Plot")
plt.show()
```

```python
plt.bar(df["Name"], df["Score"])
plt.xlabel("Student Name")
plt.ylabel("Score")
plt.title("Student Scores Bar Chart")
plt.show()
```
